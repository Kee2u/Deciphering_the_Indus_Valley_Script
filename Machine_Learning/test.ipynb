```python
# list of all words (includes duplicates)
all_words = features_df['form'].tolist()

prev_key = -1
# indexes through every row
for word_idx in range(len(all_words)):
    # if previous key matches row key, same word
    if(prev_key == features_df['key'][word_idx]):
        # iterate for loop, move to next row
        continue
    # finished looking at word, move to new word and conduct morpheme search
    else:
        prev_key = features_df['key'][word_idx]

# Morpheme search
    
    for morph_idx in range(len(morpheme2)):
        try:
            # gets starting index of morpheme in word
            str_idx = all_words[word_idx].index(morpheme2[morph_idx])
            # identify beginning of morpheme
            features_df['morpheme boundary'][word_idx + str_idx] = 1
            # identify end of morpheme
            features_df['morpheme boundary'][word_idx + str_idx + len(morpheme2[morph_idx])-1] = 2
        except:
            continue
```
